# ROOT to CSV Conversion for ML and other applications

```python
import ROOT
from root_numpy import root2array, tree2array
from root_numpy.testdata import get_filepath
import pandas as pd 
import csv 
```

Read the input file and create an array using root_numpy's tree2array

```python
inputr = "~/Desktop/Research/ROOTtutor/tree100x.root"
newfile = ROOT.TFile(inputr)
gree = newfile.Get("T")
arraynew = tree2array(gree, branches = ['Events', 'photon1', 'photon2'])
```

If we print out arraynew it lists out each entry where branch.leaf are columns of the array. 
Next, we use pandas to write it into a csv file. 

```python
df = pd.DataFrame(arraynew)
df.to_csv("file_path.csv")
```

To add an optional header, we can do the following: 

```python
with open('try.csv', 'wb') as outcsv:
    writer = csv.DictWriter(outcsv, fieldnames = ["Events", "photon1_pt", "photon1_eta",
                                                 "photon1_phi", "photon2_pt",
                                                 "photon2_eta", "photon2_phi"])
    writer.writeheader()

df = pd.DataFrame(arraynew)
df.to_csv("photontree_sample.csv")
```
